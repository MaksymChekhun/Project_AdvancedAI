```python
#import data
import pandas as pd
dataset = pd.read_csv('data.csv')
dataset.head()
```

```python
dataset.info()
```

```python
#One hot encoding Market Category
dataset['Market Category'] = dataset['Market Category'].fillna('Unknown')
dataset = pd.get_dummies(dataset, columns=['Market Category'])
```

```python
#drop Market Category,Number of Doors, popularity
dataset.drop(['Number of Doors','Popularity'],axis=1,inplace=True)
```

```python
dataset.info()
```

```python
#Normalize the numbers
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
numeric_columns = ['Year', 'Engine HP', 'Engine Cylinders', 'highway MPG', 'city mpg', 'MSRP']
dataset[numeric_columns] = scaler.fit_transform(dataset[numeric_columns])
dataset.head()
```

```python
categorical_cols = ['Make', 'Engine Fuel Type', 'Transmission Type', 'Driven_Wheels', 'Vehicle Size', 'Vehicle Style']
dataset = pd.get_dummies(dataset, columns=categorical_cols)
```

```python
dataset.head()
```

```python
# Set the 'Model' column as the index so it's not included in the math
features_df = dataset.set_index('Model')

# Ensure all remaining data is numeric (convert those 'False'/'True' to 0/1 if they aren't already floats)
features_df = features_df.astype(float)
```

```python
# 2. Fill missing numerical values with the median
numerical_cols_with_nans = ['Engine HP', 'Engine Cylinders'] 

for col in numerical_cols_with_nans:
    # Check if the column exists to avoid errors
    if col in dataset.columns:
        median_value = dataset[col].median()
        dataset[col] = dataset[col].fillna(median_value)
```

```python
categorical_cols_with_nans = ['Engine Fuel Type', 'Number of Doors']

for col in categorical_cols_with_nans:
    if col in dataset.columns:
        # Get the most frequent value
        mode_value = dataset[col].mode()[0]
        dataset[col] = dataset[col].fillna(mode_value)
```

```python
print(dataset.isnull().sum())
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Calculate the similarity matrix for all cars against all cars
# This creates a massive grid where row i and column j is the similarity score between car i and car j
# Fill any remaining missing values before computing similarity
features_df = features_df.copy()
features_df = features_df.fillna(features_df.median(numeric_only=True)).astype(float)

cosine_sim_matrix = cosine_similarity(features_df)
```

```python
indices = pd.Series(features_df.index)
```

```python
def recommend_cars(model_name, cosine_sim=cosine_sim_matrix, top_n=5):
    # 1. Get the index of the target car
    idx = indices[indices == model_name].index[0]

    # 2. Get the similarity scores and sort them (highest to lowest)
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # NEW: Keep track of unique models we've found
    unique_recommendations = []
    seen_models = set()
    
    # 3. Loop through the scores
    for i in sim_scores:
        car_model = indices[i[0]] # Get the actual name of the car
        score = i[1]              # Get the similarity score
        
        # Check if it's NOT the car we searched for, AND we haven't recommended it yet
        if car_model != model_name and car_model not in seen_models:
            unique_recommendations.append((car_model, score))
            seen_models.add(car_model)
            
        # Stop once we have exactly top_n unique cars
        if len(unique_recommendations) == top_n:
            break
            
    return unique_recommendations
```

```python
# Example 1: The Sports Car Enthusiast
print("Recommendations for a '1 Series M' (Compact, High HP, Coupe, Manual):")
print(recommend_cars('1 Series M'))

# Example 2: The Family Driver (Find a Minivan or large SUV in your dataset and test it)
print("Recommendations for a 'Odyssey':")
print(recommend_cars('Odyssey'))

# Example 3: The Commuter (Find an efficient sedan and test it)
print("Recommendations for a 'Corolla':")
print(recommend_cars('Corolla'))
```

```python

# Reload the original data only for displaying readable car specifications.
# The similarity matrix is still the one computed above.
original_specs = pd.read_csv('data.csv').reset_index().rename(columns={'index': 'row_id'})

spec_cols = [
    'row_id', 'Make', 'Model', 'Year', 'Vehicle Style', 'Vehicle Size',
    'Engine HP', 'Engine Cylinders', 'highway MPG', 'city mpg', 'MSRP'
]

# Create a readable unique label so cars with the same model name are not confused.
original_specs['car_label'] = original_specs.apply(
    lambda row: f"{row['row_id']}: {row['Make']} {row['Model']} ({row['Year']})",
    axis=1
)

print("Example car labels to use as unambiguous input:")
display(original_specs[['car_label', 'Vehicle Style', 'Vehicle Size', 'Engine HP', 'highway MPG', 'city mpg', 'MSRP']].head(10))
```

```python

def find_car_rows(model_name, make=None, year=None, limit=10):
    # Find exact dataset rows for a model. This avoids ambiguity when many rows share the same model name.
    matches = original_specs['Model'].str.contains(model_name, case=False, na=False)
    if make is not None:
        matches &= original_specs['Make'].str.contains(make, case=False, na=False)
    if year is not None:
        matches &= original_specs['Year'].eq(year)
    return original_specs.loc[matches, ['car_label'] + spec_cols].head(limit)


def recommend_cars_with_specs(row_id, cosine_sim=cosine_sim_matrix, top_n=5):
    # Recommend similar cars using row_id instead of only model name.
    # This  uses the existing cosine similarity matrix from the previous cells.
    sim_scores = list(enumerate(cosine_sim[row_id]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    recommendations = []
    seen_models = set()
    target_model = original_specs.loc[row_id, 'Model']

    for candidate_row_id, score in sim_scores:
        if candidate_row_id == row_id:
            continue

        candidate_model = original_specs.loc[candidate_row_id, 'Model']
        if candidate_model == target_model or candidate_model in seen_models:
            continue

        seen_models.add(candidate_model)
        recommendations.append((candidate_row_id, score))

        if len(recommendations) == top_n:
            break

    result = original_specs.loc[[row for row, _ in recommendations], ['car_label'] + spec_cols].copy()
    result.insert(0, 'similarity', [score for _, score in recommendations])
    return result

print("Find possible rows for Honda Odyssey:")
display(find_car_rows('Odyssey', make='Honda', limit=5))

# Use one exact Odyssey row as input.
odyssey_row_id = int(find_car_rows('Odyssey', make='Honda', limit=5).iloc[3]['row_id'])
print(f"Recommendations for row {odyssey_row_id}:")
display(recommend_cars_with_specs(odyssey_row_id, top_n=5))
```

```python

def evaluate_recommender(test_row_ids, top_n=5):
    # Simple evaluation: check whether recommendations share style/size and have similar specs.
    rows = []
    for row_id in test_row_ids:
        target = original_specs.loc[row_id]
        recs = recommend_cars_with_specs(row_id, top_n=top_n)

        rows.append({
            'input_car': target['car_label'],
            'same_vehicle_style_rate': (recs['Vehicle Style'] == target['Vehicle Style']).mean(),
            'same_vehicle_size_rate': (recs['Vehicle Size'] == target['Vehicle Size']).mean(),
            'avg_price_difference': (recs['MSRP'] - target['MSRP']).abs().mean(),
            'avg_hp_difference': (recs['Engine HP'] - target['Engine HP']).abs().mean(),
            'avg_highway_mpg_difference': (recs['highway MPG'] - target['highway MPG']).abs().mean(),
            'avg_city_mpg_difference': (recs['city mpg'] - target['city mpg']).abs().mean(),
            'avg_similarity': recs['similarity'].mean(),
        })

    return pd.DataFrame(rows)

# Five example cars for evaluation.
test_row_ids = [
    int(find_car_rows('Odyssey', make='Honda', limit=2).iloc[1]['row_id']),
    int(find_car_rows('Corolla', make='Toyota', limit=1).iloc[0]['row_id']),
    int(find_car_rows('Mustang', make='Ford', limit=1).iloc[0]['row_id']),
    int(find_car_rows('F-150', make='Ford', limit=1).iloc[0]['row_id']),
    int(find_car_rows('911', make='Porsche', limit=1).iloc[0]['row_id']),
]

evaluation_table = evaluate_recommender(test_row_ids, top_n=5)
display(evaluation_table)

print("Average evaluation scores:")
display(evaluation_table.drop(columns=['input_car']).mean().to_frame('average'))
```
